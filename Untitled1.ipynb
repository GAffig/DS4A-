```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Load the data into a pandas dataframe
data = pd.read_csv('yrs_lost_econ_data_viz.csv')

# Convert categorical variables 'state' and 'region' into numerical variables
le = LabelEncoder()
data['state'] = le.fit_transform(data['state'])
data['region'] = le.fit_transform(data['region'])

# Split the data into training and testing sets
X = data.drop(columns=['total_yrs_life_lost', 'avg_yrs_life_lost'])
y = data['total_yrs_life_lost']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train a Random Forest Regressor
#regressor = RandomForestRegressor(n_estimators=100, random_state=0)
regressor = RandomForestRegressor(n_estimators=100, random_state=0, max_depth=3)

regressor.fit(X_train, y_train)

# Test the accuracy of the model on the test data
score = regressor.score(X_test, y_test)
print("Accuracy: ", score)
```
